```python
from transformers import LayoutLMv3Processor, LayoutLMv3ForTokenClassification
from PIL import Image
import pytesseract
import pandas as pd
import torch
import os
```

```python
# Puedes probar con "microsoft/layoutlmv3-base" o "microsoft/layoutlmv3-large"
processor = LayoutLMv3Processor.from_pretrained("microsoft/layoutlmv3-base", apply_ocr=True)
model = LayoutLMv3ForTokenClassification.from_pretrained("microsoft/layoutlmv3-base")
```

```python
def ocr_with_layoutlmv3(image_path):
    image = Image.open(image_path).convert("RGB")
    # El processor hará OCR y extraerá información de layout
    encoding = processor(image, return_tensors="pt")
    outputs = model(**encoding)
    tokens = processor.tokenizer.convert_ids_to_tokens(encoding["input_ids"][0])
    labels = outputs.logits.argmax(-1).squeeze().tolist()
    words = encoding.words[0]
    boxes = encoding.boxes[0].tolist()
    ocr_text = pytesseract.image_to_string(image)
    return tokens, labels, words, boxes, ocr_text
```

```python
import pytesseract

def extract_row_by_dia_de_pago(image_path, dia_de_pago="31.12.2017"):
    image = Image.open(image_path).convert("RGB")
    # OCR con coordenadas
    ocr_data = pytesseract.image_to_data(image, output_type=pytesseract.Output.DATAFRAME)

    # Busca la fila con el valor exacto de DIA DE PAGO
    target_indices = ocr_data[ocr_data['text'].astype(str).str.strip() == dia_de_pago].index.tolist()
    if not target_indices:
        print(f"No se encontró '{dia_de_pago}' en {os.path.basename(image_path)}")
        return None
    row_idx = target_indices[0]
    # Toma la coordenada Y de la fila objetivo
    y_target = ocr_data.loc[row_idx, 'top']
    # Toma todas las palabras que caen en la misma línea (tolerancia de +-5 píxeles)
    same_row = ocr_data[(ocr_data['top'] >= y_target - 5) & (ocr_data['top'] <= y_target + 5)]
    # Agrupa palabras por columnas (usando X, de izquierda a derecha)
    same_row = same_row.sort_values('left')
    texts = same_row['text'].tolist()
    return texts

# Ejemplo de uso
folder_path = "/content/drive/MyDrive/DEUDA_PREVISIONAL/PDFS/2017"
images = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.lower().endswith('.png')]

resultados = []
for img in images:
    fila = extract_row_by_dia_de_pago(img, dia_de_pago="31.12.2017")
    if fila:
        resultados.append({"archivo": os.path.basename(img), "fila": fila})

df = pd.DataFrame(resultados)
print(df)
```
